### Importing libraries

```python
import pandas as pd     #pandas: 파이썬 데이터 분석 프로그램/엑셀파일
```

### Loading csv file

.csv = 엑셀파일st, ','로 row안 데이터 구분

1. csv파일로 dataframe 만들기

```python
df = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv", sep=",")
df    #data frame = pandas 안의 read_csv함수로 해당 csv파일을 읽어서 '변수'df에 저장
```

```python
print(df.size) #셀의 개수 17000*9
print(df.ndim) #2차원 rows*column 
print(df.shape)
```

```python
df.describe()     #기본적인 통계정보들을 나타내줌
```


### Saving to CSV

현재는 변수에 적힌 메모리 상태, 이 메모리를 파일로 저장하는 법

df를 다시 csv로

```python
df.to_csv('california_housing_train.csv')
```

### Constructing a DataFrame using dict

2. dict파일로 직접 입력하여 dataframe 만들기

```python
# Some input data - a small sample of the iris dataset


#dict = 5개의 아이템, key:value
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],    #첫번째 column
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)
df 
```

### Constructing a DataFrame using pd.Series

3. series 함수 이용하여 dataframe 만들기

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```

### Changing column names

이름 변경하기

```python
df.columns    #column명
```

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

### Accessing columns

일부만 가져오고 싶을 때

```python
# Access one column 
df.loc[:, 'sepal_length']   #row행(전체':'),column렬
```

```python
# Access one column (short-hand)    #더 간단히?
df['sepal_length']
```

```python
# Access one column (shortest-hand)
df.sepal_length
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]   #리스트로 적기
```

### Accessing rows

row정보 가져오기

```python
# Access one row 
df.loc[3, :]
```

```python
# Access two rows
df.loc[[3, 0], :]
```

```python
# Access a range of rows
df.loc[2:5, :]
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]   #이게 유용하게 쓰이겠지
```

### Accessing a cell

하나의 cell가져오기

```python
df.loc[:, 'sepal_length'][4]    #통으로 가져온 하나의 column 중 4번째꺼
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns

```python
df['new_column'] = 1    #column access방법 + =1
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

```python
df = df.drop('new_column', axis=1) # axis=1: column (sepal_length	sepal_width	petal_length	petal_width	SPECIES) #axis=0: 행축 (01234
#마지막에 df에 저장하는 것까지 해야겠지?
df
```

### Adding/Removing new rows

```python
df.loc[5, :] = ''     #=''을 안했다면 오류 메시지가 떴겠지만, 이로써 5번째 행이 추가됨
df
```

```python
df = df.drop(5, axis=0) # axis=0: row #행축에서 5인지 열축에서 5인지
df
```
